```python
import joblib
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

```python
data = joblib.load('results.pkl')
```

```python
mean_ls = np.array([np.mean([r['rewards'].sum() for r in x])for x in data])
std_ls = np.array([np.std([r['rewards'].sum() for r in x])for x in data])
```

```python
x = list(range(1, 115, 2))
plt.plot(x, mean_ls, label = 'Average total rewards')
plt.fill_between(x,mean_ls-std_ls, mean_ls+std_ls, alpha = 0.3)
# plt.legend()
plt.xlabel('Episode #')
plt.ylabel('Average total rewards')
plt.savefig('/Users/zhiyuanyao/Desktop/fig1.png')
```
